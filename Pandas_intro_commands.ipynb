```python
# prompt: can ypu import basic library pandas in this workbook

import pandas as pd

# Assuming 'Salaries.csv' is in the current working directory or provide the full path
try:
    df = pd.read_csv('Salaries.csv')
    # Display the DataFrame
    print(df)
    print(df.head())
    print(df.tail())
    print(df.columns)
    print(df.shape)
    print(df.dtypes)
    print(df.info())
    pd.set_option('display.max_rows', None)
    print(df)
    print(df.isnull())
    print(df.isnull().sum())
    print(df.describe())
    print(df.describe(include='all'))
    print(df['rank'].value_counts())
    print(df.mean(numeric_only=True))
    print(df['phd'].mean())
    print(df['phd'].median())
    print(df['salary'].median())
    print(df['salary'].mean())
    print(df['salary'].mode())
    print(df[['phd', 'salary']].mean())
    print(df['salary'].var())
    print(df['salary'].std())
    print(df['salary'].skew())
    print(df['salary'].kurtosis())
    print(df['salary'].max() - df['salary'].min())
    print(df['salary'].max())
    print(df['salary'].min())
    print(df['salary'].count())
    print(df['rank'].value_counts())

except FileNotFoundError:
    print("Error: 'Salaries.csv' not found. Please ensure the file exists in the current directory or provide the correct path.")
except pd.errors.EmptyDataError:
    print("Error: 'Salaries.csv' is empty.")
except pd.errors.ParserError:
    print("Error: There was an issue parsing 'Salaries.csv'.  Check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
